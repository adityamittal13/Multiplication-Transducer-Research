```python
# Testing Matplotlib Visualization

def createPitch():
    
    #Create figure
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)

    #Pitch Outline & Centre Line
    plt.plot([0,0],[0,90], color="blue")
    plt.plot([0,130],[90,90], color="orange")
    plt.plot([130,130],[90,0], color="green")
    plt.plot([130,0],[0,0], color="red")
    plt.plot([65,65],[0,90], color="pink")
    
    #Left Penalty Area
    plt.plot([16.5,16.5],[65,25],color="orange")
    plt.plot([0,16.5],[65,65],color="black")
    plt.plot([16.5,0],[25,25],color="red")
    
    #Right Penalty Area
    plt.plot([130,113.5],[65,65],color="pink")
    plt.plot([113.5,113.5],[65,25],color="green")
    plt.plot([113.5,130],[25,25],color="blue")
    
    #Left 6-yard Box
    plt.plot([0,5.5],[54,54],color="red")
    plt.plot([5.5,5.5],[54,36],color="green")
    plt.plot([5.5,0.5],[36,36],color="blue")
    
    #Right 6-yard Box
    plt.plot([130,124.5],[54,54],color="pink")
    plt.plot([124.5,124.5],[54,36],color="green")
    plt.plot([124.5,130],[36,36],color="red")
    
    #Prepare Circles
    centreCircle = plt.Circle((65,45),9.15,color="green",fill=False)
    centreSpot = plt.Circle((65,45),0.8,color="black")
    leftPenSpot = plt.Circle((11,45),0.8,color="pink")
    rightPenSpot = plt.Circle((119,45),0.8,color="red")
    
    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)
    
    #Prepare Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="red")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="red")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
    
    #Tidy Axes
    plt.axis('off')
    
    #Display Pitch
    plt.show()
    
createPitch()
```

```python
base = int(input("What base do you want to work in? "))
multiplier = int(input("What number do you want to multiply by? "))

missing_numbers = []
number = int(input("Which number do you want to exclude? If you want to stop inputting numbers, press -1. "))
while (number != -1):
    missing_numbers.append(number)
    number = int(input("Which number do you want to exclude? If you want to stop inputting numbers, press -1. "))

# (carryFrom, carryTo) --> (read, write)
multiplication_dict = {}


print("(carryFrom, carryTo) --> (read, write)")


# while (carry >= 0):
#     readingNumber = base - 1
#     print(carry)
#     while (readingNumber >= 0):
#         write = (((readingNumber*multiplier) + carry) % base)
#         carry2 = ((carry + (readingNumber*multiplier) - write)/base)
#         print(str(write) + " " + str(carry2))

#     if (readingNumber != m and write != m):
#         print("(" + carry + "," + carry2 + "," + readingNumber + "," + write + ")")
#         print("I AM HERE!")
#     readingNumber = readingNumber -1
#     print(" ")
#     carry = carry - 1

for carryFrom in range (multiplier):
    for read in range (base):
        total = carryFrom + (multiplier*read)
        write = total % base
        carryTo = int((total - write) / base)
       
        if (not((read in missing_numbers) or (write in missing_numbers))):
            multiplication_dict[(carryFrom, carryTo)] = (read, write)

for key in multiplication_dict.keys():
    print("(" + str(key[0]) + "," + str(key[1]) + ") --> " +  
          "(" + str(multiplication_dict[key][0]) + "," + str(multiplication_dict[key][1]) + ")")
```

```python
import math
from math import cos, sin, sqrt
from matplotlib.pyplot import figure
large_circle_radius = 100
small_circle_radius = 10

def make_circle(num_circle):
    # Declaring radius of plot and radius of each of the circles
    large_circle_radius = num_circle*125
    small_circle_radius = num_circle*20
    
    # Declaring colors and linestyles
    colors = ['b', 'g', 'r', 'black', 'brown', 'salmon', 'tomato', 'darkkhaki', 'olive', 'greenyellow', 'yellow', 
             'forestgreen', 'aquaramine', 'lightseagreen', 'darkslategrey', 'midnightblue', 'slateblue', 'mediumorchid',
             'deeppink']
    linestyles = ['solid', 'dotted', 'dashed', 'dashdot', (0, (1, 10)), (0, (5, 10)), (0, (3, 10, 1, 10)), 
                 (0, (3, 10, 1, 10, 1, 10)), (0, (5, 5))]
    
    # Matplotlib functionality
    fig=plt.figure()
    fig.clear(True)
    fig.set_size_inches(15, 15)
    ax=fig.add_subplot(1,1,1)
    
    # Finding where to keep the circles
    points = []
    for i in range(num_circle):
        points.append((round((large_circle_radius * cos(math.pi*2*i/num_circle)),3),
                      round((large_circle_radius * sin(math.pi*2*i/num_circle)),3)))
    
    # Making the circles
    circle_list = []
    for i in range (num_circle):
        circle_list.append(plt.Circle((points[i][0], points[i][1]), 
                           small_circle_radius, color="black", fill=False))
    
    for i in range (num_circle):
        ax.add_patch(circle_list[i])
    
    # Adding text to the circles
    for i in range (num_circle):
        plt.text(points[i][0] - 0.15*small_circle_radius, 
                 points[i][1] - 0.15*small_circle_radius, 
                 str(i), fontsize=((small_circle_radius)/num_circle))
    
    # Creating the lines
    for key in multiplication_dict.keys():
        # Declaring first and last state
        firstState = key[0]
        lastState = key[1]
        read = multiplication_dict[key][0]
        write = multiplication_dict[key][1]
        width = num_circle*7.5
        length = num_circle*5
        
        # Declaring coordinate points
        firstPointX = points[firstState][0]
        firstPointY = points[firstState][1]
        
        lastPointX = points[lastState][0]
        lastPointY = points[lastState][1]
        
        if (firstState == lastState):
            if (firstPointX > 0 and firstPointY > 0):
                x_values = [firstPointX + (sqrt(2)*small_circle_radius/2), 1.2*small_circle_radius + firstPointX]
                y_values = [firstPointY + (sqrt(2)*small_circle_radius/2), 1.2*small_circle_radius + firstPointY]
                x1_values = [1.2*small_circle_radius + firstPointX, 1.2*small_circle_radius + firstPointX - (1*small_circle_radius/2)]
                y1_values = [1.2*small_circle_radius + firstPointY, 1.2*small_circle_radius + firstPointY + (1*small_circle_radius/2)]
                x2 = 1.2*small_circle_radius + firstPointX - (1*small_circle_radius/2)
                y2 = 1.2*small_circle_radius + firstPointY + (1*small_circle_radius/2)
                dx2 = firstPointX - x2
                dy2 = firstPointY + small_circle_radius - y2
            elif (firstPointX > 0 and firstPointY < 0):
                x_values = [firstPointX + (sqrt(2)*small_circle_radius/2), 1.2*small_circle_radius + firstPointX]
                y_values = [firstPointY - (sqrt(2)*small_circle_radius/2), -1.2*small_circle_radius + firstPointY]
                x1_values = [1.2*small_circle_radius + firstPointX, 1.2*small_circle_radius + firstPointX + (1*small_circle_radius/2)]
                y1_values = [-1.2*small_circle_radius + firstPointY, -1.2*small_circle_radius + firstPointY + (1*small_circle_radius/2)]
                x2 = 1.2*small_circle_radius + firstPointX + (1*small_circle_radius/2)
                y2 = -1.2*small_circle_radius + firstPointY + (1*small_circle_radius/2)
                dx2 = firstPointX + small_circle_radius - x2
                dy2 = firstPointY - y2
            elif (firstPointX < 0 and firstPointY < 0):
                x_values = [firstPointX - (sqrt(2)*small_circle_radius/2), -1.2*small_circle_radius + firstPointX]
                y_values = [firstPointY - (sqrt(2)*small_circle_radius/2), -1.2*small_circle_radius + firstPointY]
                x1_values = [-1.2*small_circle_radius + firstPointX, -1.2*small_circle_radius + firstPointX + (1*small_circle_radius/2)]
                y1_values = [-1.2*small_circle_radius + firstPointY, -1.2*small_circle_radius + firstPointY - (1*small_circle_radius/2)]
                x2 = -1.2*small_circle_radius + firstPointX + (1*small_circle_radius/2)
                y2 = -1.2*small_circle_radius + firstPointY - (1*small_circle_radius/2)
                dx2 = firstPointX - x2
                dy2 = firstPointY - small_circle_radius - y2
            elif (firstPointX < 0 and firstPointY > 0):
                x_values = [firstPointX - (sqrt(2)*small_circle_radius/2), -1.2*small_circle_radius + firstPointX]
                y_values = [firstPointY + (sqrt(2)*small_circle_radius/2), 1.2*small_circle_radius + firstPointY]
                x1_values = [-1.2*small_circle_radius + firstPointX, -1.2*small_circle_radius + firstPointX - (1*small_circle_radius/2)]
                y1_values = [1.2*small_circle_radius + firstPointY, 1.2*small_circle_radius + firstPointY - (1*small_circle_radius/2)]
                x2 = -1.2*small_circle_radius + firstPointX - (1*small_circle_radius/2)
                y2 = 1.2*small_circle_radius + firstPointY - (1*small_circle_radius/2)
                dx2 = firstPointX - small_circle_radius - x2
                dy2 = firstPointY - y2
            elif (firstPointY == 0):
                if (firstPointX > 0):
                    x_values = [firstPointX + (1.83*small_circle_radius/2), firstPointX + (sqrt(2)*small_circle_radius/2)+1.2*small_circle_radius]
                    y_values = [firstPointY - small_circle_radius + (sqrt(2)*small_circle_radius/2), firstPointY - small_circle_radius + (sqrt(2)*small_circle_radius/2)]
                    x1_values = [firstPointX + (sqrt(2)*small_circle_radius/2)+1.2*small_circle_radius, firstPointX + (sqrt(2)*small_circle_radius/2) + 1.2*small_circle_radius]
                    y1_values = [firstPointY - small_circle_radius + (sqrt(2)*small_circle_radius/2), firstPointY + small_circle_radius - (sqrt(2)*small_circle_radius/2)]
                    x2 = firstPointX + (sqrt(2)*small_circle_radius/2) + 1.2*small_circle_radius
                    y2 = firstPointY + small_circle_radius - (sqrt(2)*small_circle_radius/2)
                    dx2 = firstPointX + (small_circle_radius) - x2
                    dy2 = firstPointY + (0.63*small_circle_radius/2) - y2
                else:
                    x_values = [firstPointX-(1.83*small_circle_radius/2), -1.2*small_circle_radius-(sqrt(2)*small_circle_radius/2) + firstPointX]
                    y_values = [firstPointY + (sqrt(2)*small_circle_radius/2) - small_circle_radius, firstPointY + (sqrt(2)*small_circle_radius/2) - small_circle_radius]
                    x1_values = [-1.2*small_circle_radius-(sqrt(2)*small_circle_radius/2) + firstPointX, -1.2*small_circle_radius-(sqrt(2)*small_circle_radius/2) + firstPointX]
                    y1_values = [firstPointY + (sqrt(2)*small_circle_radius/2) - small_circle_radius, firstPointY - (sqrt(2)*small_circle_radius/2) + small_circle_radius]
                    x2 = -1.2*small_circle_radius-(sqrt(2)*small_circle_radius/2) + firstPointX
                    y2 = firstPointY - (sqrt(2)*small_circle_radius/2) + small_circle_radius
                    dx2 = firstPointX - (small_circle_radius) - x2
                    dy2 = firstPointY + (0.63*small_circle_radius/2) - y2
            elif (firstPointX == 0):
                if (firstPointY > 0):
                    x_values = [firstPointX + small_circle_radius - (sqrt(2)*small_circle_radius/2), small_circle_radius + firstPointX - (sqrt(2)*small_circle_radius/2)]
                    y_values = [firstPointY + (1.83*small_circle_radius/2), 1.2*small_circle_radius + firstPointY + (1.83*small_circle_radius/2)]
                    x1_values = [small_circle_radius + firstPointX - (sqrt(2)*small_circle_radius/2), -1*small_circle_radius + firstPointX + (sqrt(2)*small_circle_radius/2)]
                    y1_values = [1.2*small_circle_radius + firstPointY + (1.83*small_circle_radius/2), 1.2*small_circle_radius + firstPointY + (1.83*small_circle_radius/2)]
                    x2 = -1*small_circle_radius + firstPointX + (sqrt(2)*small_circle_radius/2)
                    y2 = 1.2*small_circle_radius + firstPointY + (1.83*small_circle_radius/2)
                    dx2 = firstPointX - (0.63*small_circle_radius/2) - x2
                    dy2 = firstPointY + (small_circle_radius) - y2
                else:
                    x_values = [firstPointX - small_circle_radius + (sqrt(2)*small_circle_radius/2),(sqrt(2)*small_circle_radius/2) - small_circle_radius + firstPointX]
                    y_values = [firstPointY - (1.83*small_circle_radius/2), -1.2*small_circle_radius + firstPointY - (1.83*small_circle_radius/2)]
                    x1_values = [(sqrt(2)*small_circle_radius/2) - small_circle_radius + firstPointX, (-1*sqrt(2)*small_circle_radius/2) + small_circle_radius + firstPointX]
                    y1_values = [-1.2*small_circle_radius + firstPointY - (1.83*small_circle_radius/2), -1.2*small_circle_radius + firstPointY - (1.83*small_circle_radius/2)]
                    x2 = (-1*sqrt(2)*small_circle_radius/2) + small_circle_radius + firstPointX
                    y2 = -1.2*small_circle_radius + firstPointY - (1.83*small_circle_radius/2)
                    dx2 = firstPointX + (0.63*small_circle_radius/2) - x2
                    dy2 = firstPointY - small_circle_radius - y2
            plt.plot(x_values,y_values, color=colors[read], linestyle = linestyles[write])
            plt.plot(x1_values, y1_values, color=colors[read], linestyle = linestyles[write])
            plt.arrow(x2, y2, dx2, dy2, head_width=width, head_length=length, color=colors[read], linestyle=linestyles[write])
        
        # Finding distance between coordinate points
        dx = lastPointX - firstPointX
        dy = lastPointY - firstPointY
        
        # Changing for better graphics
        if (dx > 0 and dy > 0): 
            firstPointX += small_circle_radius
            lastPointY -= small_circle_radius
        elif (dx > 0 and dy < 0):
            firstPointY -= small_circle_radius
            lastPointX -= small_circle_radius
        elif (dx < 0 and dy < 0):
            firstPointX -= small_circle_radius
            lastPointY += small_circle_radius
        elif (dx < 0 and dy > 0):
            firstPointY += small_circle_radius
            lastPointX += small_circle_radius
        elif (dy == 0):
            if (dx > 0):
                firstPointY += small_circle_radius
                lastPointY += small_circle_radius
            else:
                firstPointY -= small_circle_radius
                lastPointY -= small_circle_radius
        elif (dx == 0):
            if (dy > 0):
                firstPointX -= small_circle_radius
                lastPointX -= small_circle_radius
            else:
                firstPointX += small_circle_radius
                lastPointX += small_circle_radius
        
        # Change distance between coordinate points
        dx = lastPointX - firstPointX
        dy = lastPointY - firstPointY
         
        # Add arrows
        if (firstState == lastState):
            continue
        else:
#             ax.annotate("", xy=(firstPointX, firstPointY),
#                        xytext=(lastPointX, lastPointY),
#                        arrowprops=dict(arrowstyle="->"))

            plt.arrow(firstPointX, firstPointY, dx, dy, head_width=width, head_length=length, color=colors[read], linestyle = linestyles[write])
    
    print()
    print("Legend:")
    reduce = 0
    for index in range(base):
        if (index in missing_numbers):
            reduce += 0.1*large_circle_radius
            continue
        text = "Write = " + str(index) + ":"
        plt.text(-1*large_circle_radius, large_circle_radius*1.5+(0.1*large_circle_radius*index) - reduce, text, fontsize=(12))
        plt.plot([0, small_circle_radius*5], [large_circle_radius*1.5+(0.1*large_circle_radius*index) - reduce, large_circle_radius*1.5+(0.1*large_circle_radius*index) - reduce], color = colors[index], linestyle = "solid")
    
    starting_y_value = large_circle_radius*1.5+(0.1*large_circle_radius*(base+1))
    reduce = 0
    for index in range(base):
        if (index in missing_numbers):
            reduce += 0.1*large_circle_radius
            continue
        text = "Read = " + str(index) + ":"
        plt.text(-1*large_circle_radius, starting_y_value+(0.1*large_circle_radius*index)-reduce, text, fontsize=(12))
        plt.plot([0, small_circle_radius*5], [starting_y_value+(0.1*large_circle_radius*index)-reduce, starting_y_value+(0.1*large_circle_radius*index)-reduce], color = "black", linestyle = linestyles[index])
    
    plt.axis('off')
    plt.plot()


make_circle(multiplier)
```
